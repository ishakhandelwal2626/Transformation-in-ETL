Q1. Define Data Transformation in ETL and explain why it is important.

Answer -  Transformation is the stage in the ETL process where raw, unorganized, inconsistent, or incomplete data is systematically converted into clean, standardized, and analysis-ready formats.

● It involves applying a set of business rules, calculations, cleaning techniques, and restructuring methods to ensure that the data is meaningful, accurate, and usable for decision-making.

Importance -

● It ensures Data quality as raw data contains typing mistakes, incomplete entries, duplicate values, etc. So transformation removes these issues through cleaning, standardisation, normalisation.

● It prepares datasets for analysis as transformation ensures that data is properly structured, typed and ready for visualisation.

● It enables machine learning as it helps in converting text to numbers (encoding), scale values, etc.

● It reduces errors in reporting.

Q3. What is the difference between Normalization and Standardization?

Answer -** Normalization**

Normalization is the process of scaling data values to a fixed range, usually 0 to 1.

Purpose:

- To bring different features into the same scale.

- Prevent features with large values dominating smaller ones.

Example:
Suppose the values of Age range from 20 to 60.

For Age = 40: (40−20)/(60−20)=20/40=0.5

So 40 becomes 0.5 after normalization.

Where it is used:

- Machine learning models like KNN, Neural Networks, Gradient Descent algorithms.

- When data has different ranges.

**Standardization**

Standardization transforms data so that it has a mean of 0 and a standard deviation of 1.

Purpose:

- To make data follow a standard normal distribution.

- Helps algorithms that assume normally distributed data.

Example:
If:
Mean salary = 50,000
Standard deviation = 10,000
Salary = 60,000
(60000−50000)/10000=1

So the standardized value is 1.

Where it is used:

- Algorithms like Logistic Regression, SVM, PCA.

- When data distribution matters.

Q4. A dataset has missing values in the “Age” column. Suggest two techniques to handle this and explain when they should be used.

Answer - 1. **Mean/Median Imputation**

In this method, missing values are replaced with the mean or median of the Age column.

Mean imputation: Replace missing values with the average age of all records.

Median imputation: Replace missing values with the middle value of the age distribution.

**When to use:**

- When the number of missing values is small.

- Mean is used when the data is normally distributed.

Median is preferred when there are outliers, because it is more robust.

Example:
If ages are 20, 22, 25, 60, NA, the median (25) can replace the missing value.

2.** Deletion (Removing Missing Records)**

In this technique, rows containing missing Age values are removed from the dataset.

When to use:

- When the percentage of missing data is very small.

- When removing those rows does not affect the dataset size or analysis significantly.

Example:
If only 2–3 rows out of 1000 have missing Age values, they can be safely removed.

Q5. Convert the following inconsistent “Gender” entries into a standardized format (“Male”, “Female”): ["M", "male", "F", "Female", "MALE", "f"]

Answer - To standardize the inconsistent “Gender” values into the format “Male” and “Female”, we convert different representations to a single consistent format.
- All variations such as “M”, “male”, “MALE” are converted to “Male”.

- All variations such as “F”, “Female”, “f” are converted to “Female”.

This process is called data standardization and is done during the data cleaning or transformation stage of ETL to ensure consistency in categorical data for analysis.

Q8. Explain why Z-score Standardization is preferred over Min-Max Scaling when outliers exist.

Answer - When outliers exist in a dataset, Z-score Standardization is generally preferred over Min-Max Scaling because it is less affected by extreme values.

1. Effect of Outliers on Min-Max Scaling

- Min-Max Scaling uses the minimum and maximum values of the dataset.

- If an outlier is extremely large or small, it changes the min or max drastically.

- As a result, most of the other values get compressed into a very small range, which can distort the data distribution.

Example:
If ages are 20, 22, 23, 25, 100, the value 100 (outlier) will stretch the range, causing normal values to become very close to 0 after scaling.

2. Why Z-score is Better

- Z-score uses mean and standard deviation, not the minimum or maximum.

- Although outliers affect the mean and standard deviation slightly, they do not compress the entire dataset the way Min-Max Scaling does.

- The data is transformed based on how far each value is from the mean, making it more stable when outliers exist.
